```python
import pandas as pd
import numpy as np
from scipy import stats
import os
import sys
```

```python
CPI_filelist = os.listdir('CPI')
singlefile = CPI_filelist[-1]
cpi_t = pd.read_csv('CPI\\' + singlefile, encoding = "ISO-8859-1", engine='python', sep=';')
EU_countries = cpi_t[cpi_t['Region'] == 'EU']['Country']
EU_countries = EU_countries.to_list()
t = 2000
Dataset = pd.DataFrame([], columns = ['country','year', 'cpi'])
i = 0
print('Files:')
for file in CPI_filelist:
    print(file)
    if t < 2012:
        cpi_t = pd.read_csv('CPI\\' + file)
        for country in cpi_t['country']:
            year_list = list([])
            data_temp = cpi_t[cpi_t['region'].str.contains('EU')]
            value = data_temp[data_temp['country'] == country]['score']
            if len(value) > 0:
                Dataset.loc[i] = [country,t,str(value[value.index[0]])]
                i += 1
    elif t == 2012:
        cpi_t = pd.read_csv('CPI\\' + file, encoding = "ISO-8859-1", engine='python', sep=';')
        for country in cpi_t['Country Rank']:
            year_list = list([])
            data_temp = cpi_t[cpi_t['Country'] == 'EU']
            value = data_temp[data_temp['Country Rank'] == country]['Territory']
            if len(value) > 0:
                Dataset.loc[i] = [country,t,str(value[value.index[0]])]
                i += 1
    elif t == 2013 or t == 2014 or t == 2015 or t == 2017 or t == 2018 or t == 2019:
        cpi_t = pd.read_csv('CPI\\' + file, encoding = "ISO-8859-1", engine='python', sep=';')
        for country in cpi_t['Country']:
            year_list = list([])
            data_temp = cpi_t[cpi_t['Region'] == 'EU']
            value = data_temp[data_temp['Country'] == country]['CPI']
            if len(value) > 0:
                Dataset.loc[i] = [country,t,str(value[value.index[0]])]
                i += 1
    elif t == 2016:
        cpi_t = pd.read_csv('CPI\\' + file, encoding = "ISO-8859-1", engine='python', sep=';')
        for country in cpi_t['Country']:
            year_list = list([])
            value = cpi_t[cpi_t['Country'] == country]['CPI']
            if len(value) > 0 and country in EU_countries:
                Dataset.loc[i] = [country,t,str(value[value.index[0]])]
                i += 1
    t += 1
print('Output:')
Dataset
```

```python
years = ['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009',
         '2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
cols = years
cols.insert(0,'country')
Dataset_CPI = pd.DataFrame([], columns = cols)
for country in EU_countries:
    time_series = {}
    write_country = True
    for year in years:
        nan = True
        for index, row in Dataset.iterrows():
            if country in row['country'] and year in str(row['year']):
                if write_country == True:
                    time_series.update({'country' : row['country']})
                    write_country = False
                if int(year) > 2011:
                    time_series.update({year : row['cpi']})
                else:
                    time_series.update({year : str(float(row['cpi'])*10)})
                nan = False
        if nan == True:
            time_series.update({year : 'NaN'})
    print('-', end ='')
    Dataset_CPI.loc[len(Dataset_CPI)] = time_series
Dataset_CPI
```

```python
Dataset_CPI.to_csv('CPI.csv', index_label = 'index')
```
